```python
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
# Extract year
date['date_year']=date['date'].dt.year
```

```python
date.head()
```

```python
# extract month
date['date_month']=date['date'].dt.month
```

```python
date.head()
```

```python
# extract day
date['date_day']=date['date'].dt.day
```

```python
date.head()
```

```python
# extract the month name
date['date_month_name']=date['date'].dt.month_name()
```

```python
date.head()
```

```python
# extract day of week
date['date_day_week']=date['date'].dt.dayofweek
date.head()
```

```python
# day of week -name
date['date_day_name']=date['date'].dt.day_name()
date.head()
```

```python
# is weekend
date['date_is_weekend']=np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# extract week of year
date['date_week_year']=date['date'].dt.week
date.head()
```

```python
#extract quarter
date['quarter']=date['date'].dt.quarter
date.head()
```

```python
#extract semester
date['semester']=np.where(date['quarter'].isin([1,2]),1,2)
date.head()
```

```python
# extract time elapse betweem dates
import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
# months passed
np.round((today-date['date'])/np.timedelta64(1,'D')% 30)
```

```python
np.round((today-date['date'])/np.timedelta64(1,'D')/30)
```
