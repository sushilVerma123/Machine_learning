```python
post=[]
rating=[]
company_name=[]
expirence=[]
place=[]
salary_p_tag=[]
for j in range(100):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',}
    webpage=requests.get(f'https://www.ambitionbox.com/jobs/search?designation=application-developer&rid=naukri_130424907166&page={j}',headers=headers).text
    soup=BeautifulSoup(webpage)
    company=soup.find_all('div',class_="jobInfoCard")
    for i in company:
        post.append(i.find('a', class_="title noclick").text)
        rating.append(float(i.find('span',class_="body-small").text.strip()))
        company_name.append(i.find('p',class_="company body-medium").text.strip())
        expirence.append(i.find('p',class_="body-small-l").text.strip())
        salary_p_tag.append(i.find('p',class_="body-small-l mediumbold"))

        placediv=i.find('div',class_='entity loc')
        place.append(placediv.find('p',class_="body-small-l").text.strip())
```
