# Displot (Numerical-Categorical)

```python
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False,rug=True,color='red')
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False,rug=True,color='green')
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Fare'],hist=False,rug=True,color='red')
sns.distplot(titanic[titanic['Survived']==1]['Fare'],hist=False,rug=True,color='green')
```

# HeatMap (categorical-categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
(titanic['Survived']==0).sum()
```

```python
(titanic['Survived']==1).sum()
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),cmap='YlGnBu')
```

```python
titanic.groupby('Pclass').mean()['Survived']
```

```python
titanic.groupby('Pclass').sum()['Survived']
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='pie',autopct='%.2f')
```
