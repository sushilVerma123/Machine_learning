```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('survival.csv')
df.head()
```

# Categorical Data

# Countplot

```python
sns.countplot(x=df['Survived'])
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Pclass'])
```
